```python
cars = []
with open("cars.txt","r",encoding='utf-8') as file:
    next_is_header = False
    brand = ""
    info = []
    for line in file:
        line = line.strip()
        
        if next_is_header:
            brand = line
            info = []
        else:
            if line != "":
                info.append(line)
        
        if line == "":
            next_is_header = True
            if brand != "":
                cars.append({"brand":brand,"info":info})
        else:
            next_is_header = False

cars
            
        
```

```python
import json
f = open("cars.json", "w", encoding="utf-8")
f.write(json.dumps(cars))
f.close()
```
